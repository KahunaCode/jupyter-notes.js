## Data Visualization - turning raw data into information

http://devleague.slides.com/devleague/dataviz-d3


#### D3.js
- allows you to bind data to DOM and then apply data-driven transformations to the document
- pros: small, fast, support large datasets, interaction, animations
- cons: modern browsers (IE9+), learning curve


#### Enter, Update, and Exit ... and selections
- enter things into DOM, update things, and then exit/complete
- this is the lifecycle for D3
- Selection API methods
 - target DOM methods and update them
 - #food ... any id="foo"
 - foo  ... <foo>
 - .foo  ... any class = "foo"
 - foo = bar ... any foo="bar"
 
- d3.select(selector)
- d3.selectAll(selector)
 - let elements = d3.select('.chart')
- selection.filter()
 - can put the filter element in the () or write a function
- selection.merge()
 - merge the enter and update sections
- function argument
 - d: data value mapped to the element
 - i: index of current element
 - nodes: entire groups of elements
 - .style('fill', function(d,i,nodes){
     return
     });

#### Attrubute modification
- https://bost.ocks.org/mike/selection/
-selection.attr(name, [,value])
 - value can be a constant value
- selection.style(prop, [,value])
 - you only want to use styles with D3 if you are going to be dynamically changing the values
 - if not dynamically change styles, use CSS styling
 - .style('background-color', function(d,i,nodes){ return `rgb(${d.votes + 100}, 0 0)` });
- selection.property(name, [,value])
 - used for properties which you cannot set with .attr(). checkbox's checked property is one example
- selection.append(value)
 - e.g. find all the P tags and append a div
- selection.remove()
 - removes an element
- .html(), .classed(), .insert(), .sort(), .order(), .create() ...

#### D3 Data
- https://bost.ocks.org/mike/selection/
- top level of data must be an array
- selections return Array of DOM elements
- D3 is trying to stay close to vanilla JS, thus using arrays are not a coincidence
- let data = [1,2,3,4,5,6]
- D3 does not create elements, it takes data and puts it inside elements
- svg.selectAll("circle").data(data).enter().append('circle').attr("cx", x).attr("cy", y).attr("r", 2.5);

#### Data-Join
- .data() and .enter()
- basics are that you find the element, attach the data to the element, then udpate the attr
 - .attr('cy', (data) => { return data.y };
- exiting: this is for elements with no new data

#### SVG - scalable vector graphics
- D3 uses SVGs to create and modify graphical elements to visualize our data into something meaningful to viewers
- use SVGs to make circle, triangles, charts, etc
