## Angular

- http://devleague.slides.com/devleague/interactive-intro-angularjs
- structural framework for HTML templating and allows extending HTML syntax to render custom components
- competator to React, similar ideas but wildly different execution.
- MVC: seperate all your concerns with Model, View, Controller
 - Model is the data, like a database. 
 - View- user front end, the HTML.
 - Controller- glue btw model and controller
- MVVM - Model View ViewModel is Angular's twist. ViewModel takes place of controller
- Highly opinionated. One way to do things and if you don't, all sorts of problems occur
- React vs Angular: Library vs Framework. Both different ideas that get to the same goal
 - Angular works with MVC/MVVM, React only works with the View (Redux handles state, which is like the Model)
- AngularJS is used for versions less than 2
- Angular splits responsibilities into modules and makes it very testable
 - karma, jasmine (unit tests)
 - protractor (end to end tests)
 
#### Models
- a model is a singleton defined by service
- there is only one instance ever of that object. e.g. there will only be one User service in an application, etc
 - all parts of the application see that same service, all update the same data, and cannot make seperate ones
 - like redux, the single source of truth
- all business logic should be handled by services
- model provides an excellent way to separate data and display
- great for unit testing

#### View -- Templating 
- works just like handlebars. Angular adpoted handlebars practice b/c it is widespread
- Jade
- every View is paired with at least one controller. often one view to one controller

#### Controller
- sits btw HTML views and models. interlayer between the two and should be as **thin** as possible that marries the two together
- similar to using express to render html's with handlebars
 - CONTROLLER: router.get('/user', (req,res) =>{ db.getUser().then((user)=>{res.render('user', user) })}
 - VIEW: <div> {{name}} </div> 
- there should **not** be business logic in the controller, keep it **thin**
- should not care about presentation or DOM manipulation

#### Two Way Data Binding
- if something changes on the View or Controller, the correlating Controller and View **must** change. Something will happen on both sides. This is through **SCOPE** and both sides are affected. 
 - SCOPE is nested, there are variable levels of scope. rootScope is the mother of all scopes and all scopes can see the rootScope
- good and bad. don't have crazy this.onChangeHandler.bind(this), but sometimes you aren't aware how things are changing and this can confuse and screw with you. Nothing tells you that something is changing, whereas React you have to explicitly set the state and the dataflow is unidirectional 

#### Providers
- dependency injection- as long as you call the service, it will figure our your *dependency* and *inject* it into the component that you're making
- Types can be Values (string, number, boolean, object, etc), Constants (same as values, but immutable), Factory, Service, and Provider: all basically the same but all return a singleton
 - Service: creates Singletons
 - Factory: creates Services
 - Provider: creates Factories
 - after Angluar2, it is all Services

#### Directives
- special markers on DOM element (attribute, element name, comment, even CSS class) that signals AngularJS's HTML **compiler** to attach a specified behavior to the DOM element or transform DOM element and children
 - Essentially Angular's way of creating components
 - Directives are very hard to create. Great to work with, hard to make
 - Good follow-on is to learn how to create directives
- anything with a dollarsign is a signal for Angular to compile

#### Modules
- containers for different parts of your app
- collections of configuration and run blocks
- seperate parts of Angular App into specific boxes which you can give to users/etc depending on what they need

#### Routing
- very similar to React Routes. define a route, attach a component(react)/controller(angular) (which has a view)
