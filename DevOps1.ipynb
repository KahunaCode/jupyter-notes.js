#### Once droplet created
- ssh into the machine
- create user
 - adduser james .. then: usermod -a -G sudo james
 - useradd -G sudo -s /bin/bash deploy
- su to new user (james), mkdir .ssh 
- sudo cp /root/.ssh/authorized_keys .ssh
- sudo chown james:james (user:group) authorized_keys (file)
 - now user james owns the authorized keys
- logout of root and log back in as user
- sudo apt-get update
 - updates package manager
- sudo apt-get upgrade
 - updates all the packages
- sudo vim /etc/ssh/sshd_config
 - change the port to something, 2222, and change adminlogin to no from yes
- sudo service ssh restart
- ssh -p 2222 james@165.227.25.107
- sudo ufw status
 - ufw: uncomplicated firewall
- sudo ufw enable
- sudo ufw allow 2222
 - allows connections across port 2222
